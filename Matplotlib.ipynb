```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = colours, s=sizes, alpha=.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = ['a', 'b', 'c', 'd', 'e']
```

```python
y = np.random.rand(5)
plt.bar(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
y
```

```python
plt.barh(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y, '--b')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = '#A64D13', s=sizes, alpha=.5, marker='v')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```
